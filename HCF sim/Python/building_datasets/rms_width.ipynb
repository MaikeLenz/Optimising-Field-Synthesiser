## Gaussian Spectrum

```python
C=300 # nm/fs
```

```python
LAM1, LAM2 = 400, 2000 # spectrum end points nm
LAM0, DLAM = 800, 200 # centre wavelength and fwhm bandwidth nm
OM2, OM1 = 2 * np.pi * C/LAM1, 2 * np.pi * C/LAM2  # rad/fs
OM0, DOM = 2 * np.pi * C/LAM0, DLAM * 2 * np.pi * C/LAM0**2 # rad/fs
```

```python
OM0, DOM
```

```python
NUM_PTS = 1000
```

```python
om = np.linspace(OM1, OM2, NUM_PTS) 
```

```python
spec = np.exp(-4*np.log(2) * (om-OM0)**2/DOM**2)
```

```python
plt.plot(om, spec)
plt.show()
```

```python
def nth_moment(y, x, n):
    """ calcs nth moment of distribution y sampled on x """
    N = simps(y, x)
    return (1/N) * simps(x**n * y, x)
```

```python
nth_moment(spec, om, 2)
```

```python
def sigma(y, x):
    """ calcs sigma as defined above """
    return np.sqrt(nth_moment(y, x, 2) - nth_moment(y, x, 1)**2)
```

```python
def centre(y, x):
    """ calculates the centre of mass of distribution y sampled on x """
    return nth_moment(y, x, 1)
```

```python
sigma(spec, om)
```

```python
centre(spec, om)
```

```python
def rms_width(y, x):
    return 2 * sigma(y, x)
```

```python
dom_rms = rms_width(spec, om)
dom_rms
```

```python
DOM/dom_rms
```

## Lunar HCF Spectra....

```python
path = "HCF_I_omega.csv"
intens = []
omeg = []
first_line = True
with open(path, "r") as data:
    for line in data:
        if first_line:
            first_line = False
            continue
        o, I = line.split(",")
        intens.append(float(I))
        omeg.append(float(o))
```

```python
intens = np.array(intens[0:500])/max(intens[0:500])
omeg = np.array(omeg[0:500])/1e15
```

```python
om0 = centre(intens, omeg)
om0
```

```python
y0 = centre(omeg, intens)
y0
```

```python
dom_rms = rms_width(intens, omeg)
print(dom_rms)
```

```python
xmin = om0 - dom_rms/2
xmax = om0 + dom_rms/2
x = xmin, xmax
y = y0, y0
```

```python
x
```

```python
plt.plot(omeg, intens)
plt.plot(x, y, color = "g")
plt.title("HCF spectrum: Lunar")
plt.xlabel("freq (rad/s)")
plt.ylabel("intensity (arb.units)")
plt.show()
```
